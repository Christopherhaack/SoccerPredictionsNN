```python
import sqlite3 
import numpy as np
import matplotlib.pyplot as plt
conn = sqlite3.connect('downloads/soccer/database.sqlite')

c = conn.cursor()

def getSeasonInfo(season, league_id):
    statement = '''select stage,
    home_team_api_id,
    away_team_api_id, 
    home_team_goal, 
    away_team_goal
    from match where'''
    
    
    league_clause = " league_id = " + str(league_id)
    s_clause = " season = " +  "\"" + season + "\"" 
    
    statement += league_clause + " and " + s_clause
    c.execute(statement)
    gameInfo = c.fetchall()
    return gameInfo
def getSeasonPlot(gameInfo, title):
    results = dict()
    resList = []
    for game in gameInfo:
        if game[0] == 1:
            if game[3] == game[4]:
                results[game[2]] = [1]
                results[game[1]] = [1]
            elif game[3] > game[4]:
                results[game[1]] = [3]
                results[game[2]] = [0]
            else: 
                results[game[1]] = [0]
                results[game[2]] = [3]
        else:
            if game[3] == game[4]:
                results[game[2]].append(1)
                results[game[1]].append(1)
            elif game[3] > game[4]:
                results[game[1]].append(3)
                results[game[2]].append(0)
            else: 
                results[game[1]].append(0)
                results[game[2]].append(3)
    for item, res in results.items():
        season = []
        points = 0
        for game in res:
            points += game
            season.append(points)
        resList.append(season)

    x = range(1, 39)
    y = np.transpose(np.asarray(resList))
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel('games played')
    plt.ylabel('points')
    plt.show()
    
def main():
    seasonList = []
    league = 1729
    for i in range(2008, 2016):
        season = str(i) + "/" + str(i + 1)
        seasonList.append(season)
    for season in seasonList:
        gameInfo = getSeasonInfo(season, league)
        getSeasonPlot(gameInfo, season)
        
```

```python
main()
```
